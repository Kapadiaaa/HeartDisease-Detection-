```python
# Install required packages
!pip install pyspark seaborn matplotlib

# Import necessary libraries
from pyspark.sql import SparkSession
import random
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Initialize Spark Session
spark = SparkSession.builder.appName("ColabSpark").getOrCreate()
print(spark)

# Function to generate modified patient data
def generate_modified_patient_data(n=10000):
    data = []
    for i in range(n):
        patient = {
            "patient_id": i+1,
            "age": random.randint(20, 85),
            "gender": random.choice(["Male", "Female"]),
            "bp": f"{random.randint(95, 155)}/{random.randint(65, 105)}",
            "sugar": random.randint(75, 190),
            "cholesterol": random.randint(130, 260),
            "haemoglobin": round(random.uniform(11.5, 17.5), 1),
            "bmi": round(random.uniform(19, 34), 1)
        }
        data.append(patient)

    return pd.DataFrame(data)

# Generate and save the modified dataset
modified_df = generate_modified_patient_data(10000)
modified_df.to_csv("modified_patients.csv", index=False)
print("10,000 modified patient profiles generated!")

# Read data using Spark
df_spark = spark.read.csv("modified_patients.csv", header=True, inferSchema=True)
df_spark.show(5)

# Convert to Pandas for visualization
df_pd = df_spark.toPandas()

# --- 1. Heatmap of Correlations ---
plt.figure(figsize=(10, 5))
sns.heatmap(df_pd[['age', 'sugar', 'cholesterol', 'haemoglobin', 'bmi']].corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()

# --- 2. Donut Chart for Gender Distribution ---
plt.figure(figsize=(6, 6))
gender_counts = df_pd['gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=["lightblue", "lightpink"], wedgeprops={"edgecolor": "white"})
plt.gca().add_artist(plt.Circle((0,0), 0.5, fc='white'))  # Inner circle to make it a donut
plt.title("Gender Distribution")
plt.show()

# --- 3. Stacked Bar Chart for BMI Across Age Groups ---
df_pd['age_group'] = pd.cut(df_pd['age'], bins=[20, 35, 50, 65, 85], labels=["20-35", "36-50", "51-65", "66-85"])
bmi_categories = pd.cut(df_pd['bmi'], bins=[18, 24, 29, 35], labels=["Normal", "Overweight", "Obese"])
bmi_age_group = pd.crosstab(df_pd['age_group'], bmi_categories)

bmi_age_group.plot(kind="bar", stacked=True, figsize=(10, 5), colormap="viridis")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.title("BMI Distribution Across Age Groups")
plt.legend(title="BMI Category")
plt.show()

# --- 4. Line Chart for Average Sugar Levels by Age Group ---
avg_sugar_by_age = df_pd.groupby('age_group')['sugar'].mean()

plt.figure(figsize=(10, 5))
plt.plot(avg_sugar_by_age.index, avg_sugar_by_age.values, marker="o", linestyle="-", color="red")
plt.xlabel("Age Group")
plt.ylabel("Average Sugar Level")
plt.title("Average Sugar Level by Age Group")
plt.grid(True)
plt.show()

# --- 5. Bar Chart for Average Cholesterol by Age Group ---
avg_cholesterol_by_age = df_pd.groupby('age_group')['cholesterol'].mean()

plt.figure(figsize=(10, 5))
avg_cholesterol_by_age.plot(kind="bar", color=["blue", "green", "orange", "red"])
plt.xlabel("Age Group")
plt.ylabel("Average Cholesterol Level")
plt.title("Average Cholesterol Level by Age Group")
plt.xticks(rotation=0)
plt.show()
```
